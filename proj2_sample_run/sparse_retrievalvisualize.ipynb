```python
import matplotlib.pyplot as plt
import re

# 示例純文字
text = """
Interpolated Recall - Precision Averages:
    at 0.00       0.7052
    at 0.10       0.5615
    at 0.20       0.4400
    at 0.30       0.3406
    at 0.40       0.2877
    at 0.50       0.2391
    at 0.60       0.1891
    at 0.70       0.1238
    at 0.80       0.0447
    at 0.90       0.0173
    at 1.00       0.0062
"""

# 解析純文字
recall = []
precision = []

for line in text.strip().split("\n"):
    match = re.search(r"at (\d+\.\d+)\s+(\d+\.\d+)", line)
    if match:
        recall.append(float(match.group(1)))  # Recall 值
        precision.append(float(match.group(2)))  # Precision 值

# 繪製 Recall-Precision 曲線
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='o', linestyle='-', color='b')
plt.ylim(0, 1)
# 添加標題和標籤
plt.title('Interpolated Recall-Precision Curve', fontsize=16)
plt.xlabel('Recall', fontsize=14)
plt.ylabel('Precision', fontsize=14)
plt.grid(True)

# 顯示圖表
plt.show()
```
